## Temperature Distribution

The distribution plot shows how temperature values are distributed across the dataset.

```python
plt.figure(figsize=(8,5))

sns.histplot(dataset["Temp"], kde=True)

plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")

# Save the figure
plt.savefig("../reports/temp_distribution.png", dpi=300, bbox_inches="tight")

# Display the figure
plt.show()
```

### Observation

- Most temperature values are concentrated around the average.
- The distribution helps identify whether the feature is normally distributed or skewed.

## Temperature Box Plot

The box plot is used to identify outliers in the temperature feature.

```python
plt.figure(figsize=(6,4))

sns.boxplot(x=dataset["Temp"])

plt.title("Temperature Box Plot")

# Save the figure
plt.savefig("../reports/temp_boxplot.png", dpi=300, bbox_inches="tight")

# Display the figure
plt.show()
```

### Observation

- The box plot shows the spread of temperature values.
- Any points outside the whiskers represent potential outliers.

# Step 8: Multivariate Analysis

Multivariate analysis examines the relationship among multiple variables simultaneously.

A correlation heatmap is used to identify:

- Strong positive correlations
- Strong negative correlations
- Weak correlations
- Feature dependencies

Understanding these relationships helps in feature selection and improves model performance.

```python
plt.figure(figsize=(12,8))

correlation = dataset.corr(numeric_only=True)

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5)

plt.title("Correlation Heatmap")

plt.savefig("../reports/correlation_heatmap.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

### Observation

- The heatmap shows the correlation between numerical features.
- Features with values close to +1 have a strong positive relationship.
- Features with values close to -1 have a strong negative relationship.
- Features with values near 0 have little or no linear relationship.
- The heatmap helps identify important features before model training.
